### Utility Functions for Video Processing

- **`crop_center_square(frame)`**:
  - Crops the central square region of the given frame. This is useful for maintaining consistent dimensions when frames are not square.

- **`augment_frames(frames)`**:
  - Applies random augmentations to the provided frames:
    - **Horizontal Flip**: With a probability of 50%, frames are flipped horizontally.
    - **Brightness Adjustment**: With a probability of 50%, frames' brightness is adjusted by a random factor between 0.5 and 1.5.
    - **Rotation**: With a probability of 50%, frames are rotated by a random angle between -20 and 20 degrees.

- **`load_video(path, max_frames=0, resize=(IMG_SIZE, IMG_SIZE), augment=False)`**:
  - Loads a video from the specified path, extracts frames, and optionally applies augmentations:
    - **`path`**: The path to the video file.
    - **`max_frames`**: Maximum number of frames to extract. If set to 0, all frames are extracted.
    - **`resize`**: Tuple specifying the size to which each frame should be resized (default is `(IMG_SIZE, IMG_SIZE)`).
    - **`augment`**: Boolean flag to indicate whether to apply augmentations to the frames.
  - Converts frames from BGR to RGB and returns an array of frames.


````python
# Utility functions for processing videos
def crop_center_square(frame):
    y, x = frame.shape[0:2]
    min_dim = min(y, x)
    start_x = (x // 2) - (min_dim // 2)
    start_y = (y // 2) - (min_dim // 2)
    return frame[start_y:start_y + min_dim, start_x:start_x + min_dim]

def augment_frames(frames):
    if len(frames) == 0:
        return frames  # If frames are empty, return as is

    if np.random.rand() > 0.5:
        frames = np.flip(frames, axis=2)
    if np.random.rand() > 0.5:
        factor = np.random.uniform(0.5, 1.5)
        frames = np.clip(frames * factor, 0, 255)
    if np.random.rand() > 0.5:
        angle = np.random.uniform(-20, 20)
        M = cv2.getRotationMatrix2D((frames.shape[2] // 2, frames.shape[1] // 2), angle, 1)
        for i in range(len(frames)):
            frames[i] = cv2.warpAffine(frames[i], M, (frames.shape[2], frames.shape[1]))
    return frames

def load_video(path, max_frames=0, resize=(IMG_SIZE, IMG_SIZE), augment=False):
    cap = cv2.VideoCapture(path)
    frames = []
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame = crop_center_square(frame)
            frame = cv2.resize(frame, resize)
            frame = frame[:, :, [2, 1, 0]]  # Convert from BGR to RGB
            frames.append(frame)
            if len(frames) == max_frames:
                break
    finally:
        cap.release()
    
    frames = np.array(frames)
    if len(frames) == 0:
        print(f"Warning: No frames found for video at path: {path}")
        return frames
    
    if augment:
        frames = augment_frames(frames)
    return frames
````

### Feature Extractor Construction

- **`build_feature_extractor()`**:
  - Constructs a feature extractor model using the InceptionV3 architecture:
    - **Weights**: Initialized with pre-trained weights from ImageNet.
    - **Top Layer**: Excludes the fully connected top layer (`include_top=False`).
    - **Pooling**: Uses average pooling (`pooling="avg"`).
    - **Input Shape**: The input shape is set to `(IMG_SIZE, IMG_SIZE, 3)` for RGB images.
    - **Preprocessing**: Applies preprocessing specific to InceptionV3 using `keras.applications.inception_v3.preprocess_input`.
  - **Returns**: A Keras model with the specified feature extractor and preprocessing layers.

The feature extractor model can now be used to transform video frames into feature vectors for further processing.


````python
# Build feature extractor
def build_feature_extractor():
    feature_extractor = keras.applications.InceptionV3(
        weights="imagenet",
        include_top=False,
        pooling="avg",
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
    )
    preprocess_input = keras.applications.inception_v3.preprocess_input

    inputs = keras.Input((IMG_SIZE, IMG_SIZE, 3))
    preprocessed = preprocess_input(inputs)
    outputs = feature_extractor(preprocessed)
    return keras.Model(inputs, outputs, name="feature_extractor")

feature_extractor = build_feature_extractor()
````

### Data Preparation Functions

- **`process_video(args)`**:
  - **Purpose**: Processes a single video to extract features and generate masks for the frames.
  - **Inputs**:
    - `path`: Path to the video file.
    - `root_dir`: Directory where videos are stored.
    - `augment`: Boolean flag to indicate if data augmentation should be applied.
  - **Process**:
    - **Load Video**: Calls `load_video()` to read frames from the video.
    - **Frame Extraction**: Uses the feature extractor model to convert frames into feature vectors.
    - **Mask Creation**: Creates a mask indicating the presence of valid frames up to `MAX_SEQ_LENGTH`.
  - **Outputs**: 
    - `temp_frame_features`: Array of extracted frame features.
    - `temp_frame_mask`: Array indicating which frames are valid.

- **`prepare_all_videos(df, root_dir, augment=False)`**:
  - **Purpose**: Prepares features and masks for all videos in a DataFrame.
  - **Inputs**:
    - `df`: DataFrame containing video paths and labels.
    - `root_dir`: Directory where videos are stored.
    - `augment`: Boolean flag to indicate if data augmentation should be applied.
  - **Process**:
    - **Extract Paths and Labels**: Retrieves video paths and labels from the DataFrame.
    - **Process Videos**: Applies `process_video()` to each video path.
    - **Aggregate Results**: Collects features and masks, and formats them as arrays.
  - **Outputs**:
    - `(frame_features, frame_masks)`: Arrays of video frame features and masks.
    - `labels`: Array of video labels.


````python
# Prepare the dataset
def process_video(args):
    path, root_dir, augment = args
    paths = f"/nfs/primary/Simran/kaggle/input/ucf101-videos/{os.path.join(root_dir, path)}"
    frames = load_video(paths, augment=augment)
    frames = frames[None, ...]

    temp_frame_mask = np.zeros(
        shape=(
            1,
            MAX_SEQ_LENGTH,
        ),
        dtype="bool",
    )
    temp_frame_features = np.zeros(
        shape=(1, MAX_SEQ_LENGTH, NUM_FEATURES), dtype="float32"
    )

    for i, batch in enumerate(frames):
        video_length = batch.shape[0]
        length = min(MAX_SEQ_LENGTH, video_length)
        for j in range(length):
            temp_frame_features[i, j, :] = feature_extractor.predict(
                batch[None, j, :], verbose=0,
            )
        temp_frame_mask[i, :length] = 1

    return temp_frame_features.squeeze(), temp_frame_mask.squeeze()

def prepare_all_videos(df, root_dir, augment=False):
    video_paths = df["video_name"].values.tolist()
    labels = df["tag"].values

    results = [process_video((path, root_dir, augment)) for path in video_paths]
    frame_features, frame_masks = zip(*results)

    frame_features = np.array(frame_features)
    frame_masks = np.array(frame_masks)
    labels = np.array(labels)[..., None]

    return (frame_features, frame_masks), labels
````

````python
# Define RNN Model with DP
class RNNModel(nn.Module):
    def __init__(self, num_features, num_classes):
        super(RNNModel, self).__init__()
        self.lstm1 = DPLSTM(num_features, hidden_size=512, batch_first=True)
        self.lstm2 = DPLSTM(512, hidden_size=256, batch_first=True)
        self.lstm3 = DPLSTM(256, hidden_size=128, batch_first=True)
        self.fc1 = nn.Linear(128, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, num_classes)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x, _ = self.lstm1(x)
        x, _ = self.lstm2(x)
        x, _ = self.lstm3(x)
        x = self.fc1(x[:, -1, :])  # Use the last output
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.dropout(x)
        x = self.fc3(x)
        return x
````

# Experiment Overview: Training and Evaluating Models

In this experiment, we evaluate different approaches to training machine learning models, focusing on differential privacy (DP) and data augmentation. Here's a brief overview of the process:

## **1. Experiment Setup**

We define a function, `run_experiment`, to handle two scenarios: training with or without differential privacy.

## **2. Training with Differential Privacy**

- **Data Preparation:** Convert training data to tensors and split into training and validation sets.
- **Data Loaders:** Create loaders to batch and shuffle data.
- **Model Initialization:** Use an RNN model with differential privacy via the Privacy Engine.
- **Training:** Track loss and accuracy across epochs to monitor performance.

## **3. Training without Differential Privacy**

- **TensorFlow/Keras:** Define and compile a Sequential model.
- **Training:** Train the model, evaluate performance on a test set, and generate a classification report.

## **4. Keras Model Details**

- **Architecture:** Includes TimeDistributed, LSTM, and Dense layers for sequential data.
- **Compilation:** Use a suitable loss function and optimizer.

## **Conclusion**

We compare the effects of differential privacy on model performance and explore data augmentation. The goal is to assess the trade-offs and benefits of each approach. The provided code snippets illustrate the detailed implementation.


````python
# Function to run the experiment
def run_experiment(train_data, train_labels, test_data, test_labels, use_dp=False, use_mixup=False):
    if use_dp:
        # Convert training data to tensors
        train_features = torch.tensor(train_data[0], dtype=torch.float32)
        train_labels_torch = torch.tensor(train_labels, dtype=torch.long)
        # Split into training and validation sets
        total_size = len(train_features)
        val_size = int(0.3 * total_size)
        train_size = total_size - val_size
        train_dataset_full = TensorDataset(train_features, train_labels_torch)
        train_dataset, val_dataset = random_split(train_dataset_full, [train_size, val_size])
        # Prepare data loaders
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
        # Initialize the DP model
        model_dp = RNNModel(num_features=NUM_FEATURES, num_classes=len(label_processor.get_vocabulary()))
        model_dp = model_dp.cuda()  # Use GPU
        optimizer = torch.optim.SGD(model_dp.parameters(), lr=0.01, momentum=0.9)
        criterion = nn.CrossEntropyLoss()
        # Attach the Privacy Engine to the model and optimizer
        privacy_engine = PrivacyEngine()
        model_dp, optimizer, train_loader = privacy_engine.make_private_with_epsilon(
            module=model_dp,
            optimizer=optimizer,
            data_loader=train_loader,
            epochs=EPOCHS,
            target_epsilon=5.0,
            target_delta=1e-5,
            max_grad_norm=1.0,
        )
        # Initialize history trackers
        train_loss_history = []
        val_loss_history_dp = []
        val_accuracy_history_dp = []
        train_accuracy_history_dp = []
        # Training loop
        for epoch in range(EPOCHS):
            model_dp.train()
            epoch_loss = 0
            correct = 0
            total = 0
            for inputs, labels in train_loader:
                inputs, labels = inputs.cuda(), labels.cuda()  # Move data to GPU
                optimizer.zero_grad()
                outputs = model_dp(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
            # Calculate and log the training accuracy and loss
            average_loss = epoch_loss / len(train_loader)
            train_accuracy = correct / total
            train_loss_history.append(average_loss)
            train_accuracy_history_dp.append(train_accuracy)
            # Validation loop
            model_dp.eval()
            val_loss = 0
            correct_val = 0
            total_val = 0
            with torch.no_grad():
                for val_inputs, val_labels in val_loader:
                    val_inputs, val_labels = val_inputs.cuda(), val_labels.cuda()  # Move data to GPU
                    val_outputs = model_dp(val_inputs)
                    v_loss = criterion(val_outputs, val_labels)
                    val_loss += v_loss.item()
                    _, val_predicted = torch.max(val_outputs.data, 1)
                    total_val += val_labels.size(0)
                    correct_val += (val_predicted == val_labels).sum().item()
            # Calculate and log the validation accuracy and loss
            average_val_loss = val_loss / len(val_loader)
            val_accuracy = correct_val / total_val
            val_loss_history_dp.append(average_val_loss)
            val_accuracy_history_dp.append(val_accuracy)
            print(f'Epoch {epoch+1}/{EPOCHS}, '
                  f'Train Loss: {average_loss:.4f}, Train Acc: {train_accuracy:.4f}, '
                  f'Val Loss: {average_val_loss:.4f}, Val Acc: {val_accuracy:.4f}')
        return train_loss_history, train_accuracy_history_dp, val_loss_history_dp, val_accuracy_history_dp, model_dp
    else:
        # TensorFlow model training without DP
        seq_model = get_sequence_model()
        optimizer = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
        seq_model.compile(
            loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"]
        )
        history = seq_model.fit(
            [train_data[0], train_data[1]],
            train_labels,
            validation_split=0.3,
            epochs=EPOCHS,
        )
        _, accuracy = seq_model.evaluate([test_data[0], test_data[1]], test_labels)
        print(f"Test accuracy: {round(accuracy * 100, 2)}%")
        predictions = seq_model.predict([test_data[0], test_data[1]])
        predicted_labels = np.argmax(predictions, axis=1)
        report = classification_report(test_labels, predicted_labels)
        print("Classification Report:\n", report)
        return history, seq_model

# Get the Keras model
def get_sequence_model():
    class_vocab = label_processor.get_vocabulary()
    frame_features_input = keras.Input((MAX_SEQ_LENGTH, NUM_FEATURES))
    mask_input = keras.Input((MAX_SEQ_LENGTH,), dtype="bool")
    x = keras.layers.TimeDistributed(keras.layers.Dense(512, activation='relu'))(frame_features_input)
    x = keras.layers.TimeDistributed(keras.layers.Dense(256, activation='relu'))(x)
    x = keras.layers.LSTM(512, return_sequences=True)(x)
    x = keras.layers.LSTM(256, return_sequences=True)(x)
    x = keras.layers.LSTM(128)(x)
    x = keras.layers.Dropout(0.5)(x)
    x = keras.layers.Dense(128, activation="relu")(x)
    x = keras.layers.Dropout(0.5)(x)
    x = keras.layers.Dense(64, activation="relu")(x)
    output = keras.layers.Dense(len(class_vocab), activation="softmax")(x)
    rnn_model = keras.Model([frame_features_input, mask_input], output)
    optimizer = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
    rnn_model.compile(
        loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"]
    )
    return rnn_model
````

````python
# TensorFlow-compatible beta distribution sampling
def sample_beta_distribution_tf(size, concentration_0=0.2, concentration_1=0.2):
    gamma_1_sample = tf.random.gamma([size], concentration_1)
    gamma_2_sample = tf.random.gamma([size], concentration_0)
    return gamma_1_sample / (gamma_1_sample + gamma_2_sample)

def mix_up_tf(ds_one, ds_two, alpha=0.2):
    images_one, labels_one = ds_one
    images_two, labels_two = ds_two
    batch_size = tf.shape(images_one)[0]

    # Sample mixup coefficients
    l = sample_beta_distribution_tf(batch_size, alpha, alpha)
    x_l = tf.reshape(l, [batch_size, 1, 1, 1])
    y_l = tf.reshape(l, [batch_size, 1])

    # Convert labels to float32 for compatibility with the mixed coefficients
    labels_one = tf.cast(labels_one, tf.float32)
    labels_two = tf.cast(labels_two, tf.float32)

    # Mix images and labels
    images = images_one * x_l + images_two * (1 - x_l)
    labels = labels_one * y_l + labels_two * (1 - y_l)

    # Cast labels back to int64 if necessary
    labels = tf.cast(labels, tf.int64)

    return images, labels
````

````python
import time
import pandas as pd
import torch

# Inference time measurement
def measure_inference_time(model, data, is_pytorch=False, device=None):
    start_time = time.time()
    
    if is_pytorch:
        # Move data to the appropriate device (CPU or GPU)
        data = data.to(device) if device else data
        model = model.to(device) if device else model
        
        # Ensure all operations on GPU are complete before timing
        if device and 'cuda' in str(device):
            torch.cuda.synchronize()

        # Forward pass
        _ = model(data).detach().cpu().numpy()
        
        # Ensure all operations on GPU are complete before stopping the timer
        if device and 'cuda' in str(device):
            torch.cuda.synchronize()

    else:
        # For TensorFlow/Keras model
        _ = model.predict(data)

    end_time = time.time()
    return end_time - start_time

# Specify the device for PyTorch models (e.g., "cuda:0" for the first GPU)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Measure inference times
time_no_dp = measure_inference_time(model_no_dp, [test_data[0], test_data[1]])
time_dp = measure_inference_time(model_dp, torch.tensor(test_data[0], dtype=torch.float32), is_pytorch=True, device=device)
time_augmented = measure_inference_time(model_augmented, [test_data[0], test_data[1]])

# Create a DataFrame to display the results
inference_times_df = pd.DataFrame({
    "Model": ["No DP", "With DP", "With Augmentation"],
    "Inference Time (seconds)": [time_no_dp, time_dp, time_augmented]
})

print(inference_times_df)
````

````python
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
import numpy as np
from scipy.special import softmax  # Import softmax from scipy

# Define the plotting function for calibration curves
def plot_calibration_curve(y_true, y_prob, model_name, class_idx):
    prob_true, prob_pred = calibration_curve(y_true == class_idx, y_prob[:, class_idx], n_bins=10)
    plt.plot(prob_pred, prob_true, marker='o', label=f'{model_name} - Class {class_idx}')

# Get logits from the models (assume these are your pre-trained models)
logits_without_sim_dp = model_without_sim_dp([test_data[0], test_data[1]])
logits_sim_dp = model_with_sim_dp(torch.tensor(test_data[0], dtype=torch.float32).to(device)).detach().cpu().numpy()
logits_sim_augmentation = model_with_sim_augmentation([test_data[0], test_data[1]])

# Apply softmax to logits to get probabilities
outputs_no_dp_prob = softmax(logits_without_sim_dp, axis=1)
outputs_dp_prob = softmax(logits_sim_dp, axis=1)
outputs_augmented_prob = softmax(logits_sim_augmentation, axis=1)

# Number of classes
num_classes = outputs_no_dp_prob.shape[1]

# Plot calibration curves
plt.figure(figsize=(18, 12))
for class_idx in range(num_classes):
    plt.subplot(2, (num_classes + 1) // 2, class_idx + 1)
    plot_calibration_curve(test_labels_encoded, outputs_no_dp_prob, 'No DP', class_idx)
    plot_calibration_curve(test_labels_encoded, outputs_dp_prob, 'With DP', class_idx)
    plot_calibration_curve(test_labels_encoded, outputs_augmented_prob, 'With Augmentation', class_idx)
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.title(f'Class {class_idx}')
    plt.xlabel('Mean Predicted Probability')
    plt.ylabel('Fraction of Positives')
    plt.legend(loc='best', fontsize='small')

plt.tight_layout()
plt.show()
````
